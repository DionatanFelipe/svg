```python
import os

# CONFIGURAÇÃO / CONFIGURATION / CONFIGURACIÓN
input_folder = "/home/user/..."  # pasta com SVGs / folder with SVGs / carpeta con SVGs
output_folder = "/home/user/novo"  # pasta para salvar modificados / destination folder / carpeta de destino
target_color = "#00ff9c"  # cor que vai substituir a existente / new replacement color / color de reemplazo

# cria a pasta de saída se não existir / create output folder if it doesn't exist / crea la carpeta de salida si no existe
os.makedirs(output_folder, exist_ok=True)

# função para processar um arquivo SVG / function to process an SVG file / función para procesar un archivo SVG
def process_svg(file_path, output_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        content = f.read()

    # substitui cores específicas (ex: branco) / replace specific colors (e.g. white) / reemplaza colores específicos (ej: blanco)
    # adicione mais replace para outras cores / add more replace for other colors / agrega más replace para otros colores
    # exemplo / example / ejemplo:
    # content_new = content.replace('#ffffff', target_color).replace('#fff', target_color)

    content_new = content.replace('#ffffff', target_color)

    with open(output_path, 'w', encoding='utf-8') as f:
        f.write(content_new)

# percorre todas as pastas e arquivos SVG / iterate through all folders and SVG files / recorre todas las carpetas y archivos SVG
for root, dirs, files in os.walk(input_folder):
    # cria subpastas correspondentes na saída / create equivalent subfolders in output / crea subcarpetas equivalentes en la salida
    rel_path = os.path.relpath(root, input_folder)
    output_subfolder = os.path.join(output_folder, rel_path)
    os.makedirs(output_subfolder, exist_ok=True)

    for file in files:
        if file.lower().endswith('.svg'):
            input_file = os.path.join(root, file)
            output_file = os.path.join(output_subfolder, file)
            process_svg(input_file, output_file)
            print(f"Modificado / Modified / Modificado: {output_file}")
```
